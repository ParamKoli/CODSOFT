```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

print("✅ Libraries imported successfully!")
```

```python
movies = pd.DataFrame({
    'title': [
        'Inception', 'Interstellar', 'The Dark Knight', 'Batman Begins',
        'The Matrix', 'John Wick', 'Mad Max Fury Road', 'Avatar',
        'Harry Potter', 'The Chronicles of Narnia', 'The Lord of the Rings',
        'The Hobbit', 'Iron Man', 'Avengers', 'Captain America',
        'Spider Man', 'Thor', 'Black Panther', 'Doctor Strange',
        'Guardians of the Galaxy', 'Titanic', 'The Notebook',
        'Pride and Prejudice', 'La La Land', 'A Beautiful Mind',
        'Good Will Hunting', 'The Pursuit of Happyness', 'Forrest Gump',
        'The Shawshank Redemption', 'The Godfather'
    ],
    'tags': [
        'sci-fi thriller dreams mind-bending heist action',
        'sci-fi space time travel drama emotional',
        'action superhero crime dark thriller',
        'action superhero origin crime dark',
        'sci-fi action philosophy mind-bending dystopia',
        'action thriller assassin crime revenge',
        'action adventure dystopia survival thriller',
        'sci-fi fantasy adventure aliens action',
        'fantasy magic school adventure friendship',
        'fantasy adventure magic kids family',
        'fantasy adventure epic journey friendship',
        'fantasy adventure epic prequel journey',
        'superhero action comedy technology genius',
        'superhero action team adventure epic',
        'superhero action war patriotism adventure',
        'superhero action adventure teen comedy',
        'superhero action adventure mythology comedy',
        'superhero action adventure Africa kingdom',
        'superhero action magic multiverse sorcery',
        'superhero action comedy space adventure team',
        'romance drama historical tragedy love ship',
        'romance drama love story emotional',
        'romance drama historical love classic',
        'romance drama music jazz ambition love',
        'drama biography math genius mental illness',
        'drama genius math friendship emotional',
        'drama biography inspiration struggle success',
        'drama comedy history emotional journey life',
        'drama prison friendship hope classic',
        'drama crime mafia family power classic'
    ]
})

print("✅ Movie dataset created!")
print(f"Total movies: {len(movies)}")
print(movies[['title']].to_string())
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['tags'])
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

def get_recommendations(movie_title):
    movie_title = movie_title.strip().title()
    if movie_title not in movies['title'].values:
        print(f"❌ Movie '{movie_title}' not found! Try another movie.")
        print("\nAvailable movies:")
        for title in movies['title'].values:
            print(f"  - {title}")
        return

    idx = movies[movies['title'] == movie_title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    movie_indices = [i[0] for i in sim_scores]

    print("\n" + "="*40)
    print(f"🎬 Movies similar to '{movie_title}'")
    print("="*40)
    for i, idx in enumerate(movie_indices, 1):
        print(f"{i}. {movies['title'].iloc[idx]}")
    print("="*40)

print("✅ Recommendation engine ready!")
```

```python
get_recommendations("Inception")
get_recommendations("Harry Potter")
get_recommendations("Iron Man")
```
