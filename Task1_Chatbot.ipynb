```python
import random
import datetime

def chatbot(user_input):
    user_input = user_input.lower().strip()

    # Greetings
    if any(word in user_input for word in ["hello", "hi", "hey", "howdy", "sup"]):
        responses = [
            "Hey there! 😊 How can I help you today?",
            "Hello! Great to see you! 🌟",
            "Hi! What's on your mind?",
            "Hey! I'm here and ready to chat! 🤖"
        ]
        return random.choice(responses)

    # How are you
    elif any(word in user_input for word in ["how are you", "how r u", "you okay", "you good"]):
        return "I'm doing awesome, thanks for asking! 😄 How about you?"

    # Name
    elif any(word in user_input for word in ["your name", "who are you", "what are you"]):
        return "I'm ParamBot 🤖 — your personal AI assistant, built by Param!"

    # Time
    elif "time" in user_input:
        now = datetime.datetime.now().strftime("%I:%M %p")
        return f"The current time is {now} ⏰"

    # Date
    elif "date" in user_input or "today" in user_input:
        today = datetime.datetime.now().strftime("%A, %B %d, %Y")
        return f"Today is {today} 📅"

    # Jokes
    elif "joke" in user_input:
        jokes = [
            "Why do programmers prefer dark mode? Because light attracts bugs! 😂",
            "Why did the scarecrow win an award? Because he was outstanding in his field! 🌾",
            "I told my computer I needed a break. Now it won't stop sending me Kit-Kat ads 😅",
            "Why don't scientists trust atoms? Because they make up everything! ⚛️"
        ]
        return random.choice(jokes)

    # Weather
    elif "weather" in user_input:
        return "I can't check live weather yet, but I hope it's sunny in Mumbai! ☀️"

    # Age
    elif "age" in user_input or "how old" in user_input:
        return "I'm ageless! Born the moment you ran this code 😄"

    # Feelings - user
    elif any(word in user_input for word in ["i am sad", "i'm sad", "feeling sad", "depressed", "unhappy"]):
        return "Aww, I'm sorry to hear that 😢 Remember, tough times don't last. You've got this! 💪"

    elif any(word in user_input for word in ["i am happy", "i'm happy", "feeling good", "great", "awesome"]):
        return "That's amazing!! Keep that energy going! 🎉😄"

    # Compliments to bot
    elif any(word in user_input for word in ["good bot", "nice", "cool", "amazing", "smart"]):
        return "Aww thank you! You just made my day 🥹🤖"

    # Help
    elif "help" in user_input:
        return """I can help you with:
😄 Greetings & small talk
⏰ Current time & date
😂 Jokes
💬 General conversation
Just type anything and I'll try my best!"""

    # Thanks
    elif any(word in user_input for word in ["thank", "thanks", "ty", "thx"]):
        return "You're welcome! Always happy to help 😊"

    # Bye
    elif any(word in user_input for word in ["bye", "goodbye", "exit", "quit", "see you"]):
        return "Goodbye! Come back soon! 👋😊"

    # Creator
    elif any(word in user_input for word in ["who made you", "who created you", "your creator"]):
        return "I was created by Param as part of the CodSoft AI Internship! 🚀"

    # Favorite things
    elif "favorite color" in user_input:
        return "I love the color blue 💙 — like the sky and the ocean!"

    elif "favorite food" in user_input:
        return "I'd love to try pizza if I could eat! 🍕"

    # Fallback
    else:
        fallbacks = [
            "Hmm, I'm not sure about that 🤔 Try asking something else!",
            "Interesting! But I don't have an answer for that yet 😅",
            "I'm still learning! Could you rephrase that?",
            "That's beyond my knowledge for now! Ask me something else 😊"
        ]
        return random.choice(fallbacks)

# ---- Run the chatbot ----
print("=" * 40)
print("   Welcome to ParamBot 🤖")
print("   Your personal AI Assistant!")
print("=" * 40)
print("Type 'bye' to exit\n")

while True:
    user_input = input("You: ")
    if not user_input.strip():
        continue
    response = chatbot(user_input)
    print(f"ParamBot: {response}\n")
    if any(word in user_input.lower() for word in ["bye", "exit", "quit"]):
        break
```
